```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import os

# Define the parameters
qds = ['BPHQ', 'BPTQ', 'BPSQ']
estado = ['original', 'relaxed']
positions = ['P52', 'P61']
vistas = ['z', 'y']  # z first so they go in the top row

# Initialize a list to store the cropped images
cropped_images = []
names = []

# Load and crop the images (loop starting with vista)
for vis in vistas:  # vista goes outermost
    for est in estado:
        for qd in qds:
            for pos in positions:
                file_name = f'{qd}_{pos}_{est}_{vis}.png'
                if os.path.exists(file_name):
                    print(f'Cargando imagen: {file_name}')
                    imagen = Image.open(file_name)
                    imagen_np = np.array(imagen)
                    names.append(file_name)
                    if vis == 'z':
                        x1, y1, x2, y2 = 50, 50, 950, 950
                    else:
                        x1, y1, x2, y2 = 50, 450, 950, 1200
                    imagen_recortada = imagen_np[y1:y2, x1:x2]
                    cropped_images.append(imagen_recortada)

# Create a plot with 2 rows and 6 columns
fig, axes = plt.subplots(2, 6, figsize=(15, 5))

# Add column labels (a)-(f) on top of each column
column_titles = ['(a)', '(b)', '(c)', '(d)', '(e)', '(f)']
for col in range(6):
    axes[0, col].set_title(column_titles[col], fontsize=12, weight='bold')

# Plot the images
for i, ax in enumerate(axes.flat):
    ax.imshow(cropped_images[i])
    ax.axis('off')

plt.tight_layout()
plt.subplots_adjust(top=0.88)  # Leave space for titles if needed
plt.show()
```
